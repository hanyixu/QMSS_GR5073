```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib as mlp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#A1
df = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/datasets/mtcars.csv')
```

```python
#A2
df.head(5)
```

```python
#A3
df.groupby('cyl')['mpg'].mean()
```

```python
#A4
plt.hist(df['mpg'])
plt.xlabel('mile per gallon')
plt.ylabel('Frequency')
plt.title('Frequency of MPG in the table')
plt.show()
```

```python
#A5
plt.scatter(df['wt'], df['hp'])
plt.xlabel('weight(1000 lbs)')
plt.ylabel('hourse power')
plt.title('Relationship between weight and hoursepower')
plt.show()
```
